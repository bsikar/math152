### Question 1


#### 1a


```python
n = symbols("n")
a_n = 1 + 1 / (n ** 2)
for i in range(1, 11):
    print(f"a_n = {a_n} for n={i} is {a_n.subs(n, i).evalf()}")

print("It looks like the limit goes to 1")
```

#### 1b


```python
# plot the first 30 points
x = np.arange(1, 31)
y = [a_n.subs(n, i) for i in x]
plt.plot(x, y)
plt.show()
print("The graph above is the first 30 points of the sequence")
print("It looks like the limit goes to 1")
```

#### 1c


```python
limit_ = Limit(a_n, n, oo)
print(f"The limit of the sequence is {limit_.doit()}")
```

### Question 2


#### 2a


```python
n = symbols("n", integer=True)
b = (4 / 5) ** n / n
L = limit(b, n, oo)
print("The limit as n approaches infinity of b of n is:", L)
```

#### 2b


```python
# Define the sequence
n = np.arange(1, 51)
bn = ((4 / 5) ** n) / n

# Calculate the partial sums
partial_sums = np.cumsum(bn)

# Plot the sequence and partial sums
plt.plot(n, bn, label="Sequence")
plt.plot(n, partial_sums, label="Partial Sums")
plt.xlabel("n")
plt.legend()
plt.show()
print(f"The graph above shows the sequence and partial sums for the first 50 terms")
```

#### 2c


```python
S = np.cumsum(b)
print(f"Sum of series is {S[len(S) - 1]}, which is close to {L}")
```
