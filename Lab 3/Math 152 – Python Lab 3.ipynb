### Question 1

#### 1a

```python
# find the region rotated about the x-axis
x, y = symbols('x y')
y1 = x * exp(1 - x / 2)
y2 = 2

x_intersections = (0, 2)

area = pi * integrate((y2 ** 2) - (y1 ** 2), (x, *x_intersections))

print(f"The region A about the x-axis has a volume of {area.simplify()} which is approximately {area.evalf()} units cubed.")
```

#### 1b

```python
# find the region rotated about the y-axis
# I will use shell method to find the area

h = 2 - y1
R = x

area = 2 * pi * integrate(R * h, (x, *x_intersections))
print(f"The region A about the y-axis has a volume of {area.simplify()} which is approximately {area.evalf()} units cubed.")
```

### Question 2

#### 2a

```python
y = symbols('y')

distance = y + 1
density = 1000 * 9.8
volume = (3 / 2) * (2 - y) * 5
work = integrate(density * volume * distance, (y, 0, 2))

print(f"The work required to pump all the water out of the tank is {work.simplify()} J.")
```

#### 2b

```python
n = symbols('n')
res = integrate(1000 * 9.8 * (y + 1) * ((3 / 2) * (2 - y) * 5), (y, 0, n))
work = 3E4
height = nsolve(work - res, 0)

print(f"The height of the water in the tank when the work is 30,000 J is {2 - height} m.")
```
