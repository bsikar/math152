### Question 1

#### 1a

```python
x, y = symbols('x y')

fx = -x - y ** 2 + 5 * y - 3

# plot on y [-2, 5]
plot_implicit(fx, x, (y, -2, 5))
print("The graph of x = -y^2 + 5y - 3 is shown above.")
```

#### 1b

```python
fx = -y ** 2 + 5 * y - 3
y_int = solve(fx, y)
print(f"The y-intercept's for the function x = -y^2 + 5y - 3 are at {y_int}")
```

#### 1c

```python
vol = pi * integrate(fx ** 2, (y, *y_int))
print(f"The volume of the function {fx} is {vol.simplify()} which is about {vol.evalf()} when rotated about the y-axis")
```
