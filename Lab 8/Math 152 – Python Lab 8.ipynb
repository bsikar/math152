### Question 1


#### 1a


```python
# series from 1 to inf
# of (n^50*50^n)/(n!)

# the first 5 terms
n = symbols("n")
running_sum = 0
func = (n ** 50 * 50 ** n) / (factorial(n))

for i in range(1, 6):
    x = func.subs(n, i).evalf()
    running_sum += x
    print(f"For n={i} the value is {x} and the running sum is {running_sum}")

print("The series looks like its going to some very value 2E41")
```

#### 1b


```python
n = symbols("n")
an = (n ** 50 * 50 ** n) / factorial(n)
an_plus1 = (n + 1) ** 50 * 50 ** (n + 1) / factorial(n + 1)

print(f"an={an}")
print(f"an+1={an_plus1}")

func = (an_plus1 / an).simplify()

print(f"an+1/an={func}")

lim = limit(func, n, oo).doit().evalf()

print(f"The limit is {lim}")

lim = abs(lim)

print(f"The absolute value of the limit is {lim}")

print(f"Since the limit is 0 it converges absolutely : if L < 1 then the series converges absolutely")
```

#### 1c


```python
print("THe answer to part b tells us that the series converges absolutely")
```
