```python
x = symbols("x", positive=True)
a = symbols("a", positive=True)
f = x ** 2 / (x ** 4 + a ** 2)
F = integrate(f, (x, 0, oo))
ans = solve(F - 0.1, a)
print("our a value is:", ans)
```

```python
# show that g(x) = 1/x^3 converges
# via the integral from 1 to inf
x = Symbol("x")
g = 1 / x ** 2

value = integrate(g, (x, 1, oo))

print(f"Via the integral, the value of the integral is {value}")
print("Since I got a finite value, the integral converges")
```
