### Question 1


#### 1a


```python
x = sp.symbols("x")
f = (5 + sp.sin(x)) ** (1 / 4)

lower = 0
upper = pi / 2

# Plot the function
sp.plot(f, (x, lower, upper))
print(f"The graph of f(x) = {f} is shown above.")
```

#### 1b Left Endpoint Approximation


```python
n = 200
delta_x = (upper - lower) / n
# make a list from inclusive lower to exclusive upper
# with step size delta_x
x_values = arange(lower, upper, delta_x)
y_values = [f.subs(x, x_value) for x_value in x_values]
# sum the list y_values and multiply by delta_x
# to get the area under the curve
area_left = sum(y_values) * delta_x
print(f"The area under the curve is {area_left}.")
```

### Question 2 Right Endpoint Approximation


```python
n = 200
delta_x = (upper - lower) / n
# make a list from exclusive lower to inclusive upper
# with step size delta_x
x_values = arange(lower + delta_x, upper + delta_x, delta_x)
y_values = [f.subs(x, x_value) for x_value in x_values]
# sum the list y_values and multiply by delta_x
# to get the area under the curve
area_right = sum(y_values) * delta_x
print(f"The area under the curve is {area_right}.")
```

### Question 3 Midpoint Approximation


#### 3a


```python
# start at a + delta_x and go to b - delta_x inclusive
# with step size delta_x
x_values = arange(lower + delta_x / 2, upper + delta_x / 2, delta_x)
y_values = [f.subs(x, x_value) for x_value in x_values]
# sum the list y_values and multiply by delta_x
# to get the area under the curve
area_mid = sum(y_values) * delta_x
print(f"The area under the curve is {area_mid}.")
```

#### 3b


```python
left_right_average = (area_left + area_right) / 2
print(f"The average of the left and right areas is {left_right_average}.")

print(f"Is the average of the left and right areas equal to the area under the curve using midpoint? {left_right_average == area_mid}.")
```

### Question 4 Trapezoid Approximation


#### 4a


```python

from scipy.integrate import trapz

x_values = linspace(lower, upper, n + 1)
y_values = [f.subs(x, x_value) for x_value in x_values]

result = trapz(y_values, x_values).evalf()

print(f"The area under the curve is {result}.")
```

### Question 5 Simpson's Rule


```python
import numpy as np
from scipy.integrate import simps

a = 0
b = np.pi / 2
n = 200

x = np.linspace(a, b, n)
y = (5 + np.sin(x)) ** 1 / 4
sim = simps(y, x)
print(sim)
```
